## NumPy-创建ndarray数组之array函数
- array()

```python
import numpy as np
```

```python
help(np.array)
```

### np.array()

可以通过给np.array()函数传递Python的序列对象来创建数组，如果传递的是多层嵌套的序列，将创建多维数组。

#### 使用一维列表创建

```python
list1 = [1, 3, 5, 7, 9]
np.array(list1)
```

#### 使用二维列表创建

```python
list2 = [
    [1, 3, 5, 7, 9],
    [2, 4, 6, 8, 0]
]
np.array(list2)
```

#### 使用一维元组创建

```python
tup1 = (1, 2, 3, 4, 5)
np.array(tup1)
```

#### 使用一维元组创建

```python
tup2 = (
        (1, 3 ,5, 7, 9),
        (2, 4, 6, 8, 10))
np.array(tup2)
```
